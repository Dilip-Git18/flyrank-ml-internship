I selected the content decline prediction lane because identifying pages that are likely to lose search performance can help SEO teams prioritize their work. # 2. The Question: Decision, Action, Cost of a Wrong Call

## Research Question

Can historical search and engagement metrics be used to identify content pages that are likely to decline in search performance?

## Unit of Analysis

One content page.

## Expected Output

A binary prediction:

- Declining
- Not Declining

## Decision

Which pages should an SEO team review first?

## Action

Pages predicted to decline can be prioritized for content updates, keyword optimization, or technical improvements.

## Cost of an Incorrect Recommendation

### False Positive

A page predicted to decline may actually remain stable, causing unnecessary effort.

### False Negative

A page that is actually declining may not be reviewed in time, potentially leading to reduced traffic and missed opportunities.

Machine learning can assist in prioritizing work but should be used only as decision support rather than making automatic decisions.

```python
import pandas as pd
from pathlib import Path

possible_paths = [
    Path("/content/flyrank-ml-internship/data/raw/content_refresh_anonymized.csv"),
    Path("/content/flyrank-ml-internship/content_refresh_anonymized.csv"),
    Path("../../data/raw/content_refresh_anonymized.csv"),
    Path("../data/raw/content_refresh_anonymized.csv"),
    Path("data/raw/content_refresh_anonymized.csv"),
    Path("content_refresh_anonymized.csv"),
]

dataset_path = None

for path in possible_paths:
    if path.exists():
        dataset_path = path
        break

if dataset_path is None:
    raise FileNotFoundError("Dataset not found!")

print("Dataset found at:", dataset_path)

df = pd.read_csv(dataset_path)

print("Dataset Shape:", df.shape)

display(df.head())

print("\nAverage CTR:", round(df["ctr"].mean(), 2))
print("Average Search Position:", round(df["avg_position"].replace(0, pd.NA).dropna().mean(), 2))
print("Average Engagement Rate:", round(df["engagement_rate"].mean(), 2))

print("\nRows:", len(df))
print("Columns:", len(df.columns))
```

# 4. Careful Words: What I Can and Can't Claim

This project analyzes patterns that are observed in the provided FlyRank dataset.

The machine learning model may identify relationships between historical content metrics and future content decline. These relationships are observational and should be interpreted as decision-support evidence.

This project does not prove that one variable causes another. It also cannot predict Google's ranking algorithm or guarantee future search performance.

The results should be used to help prioritize content review and optimization, while final decisions remain with human experts.
