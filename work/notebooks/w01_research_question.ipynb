# 2. The Question: Decision, Action, Cost of a Wrong Call

## Research Question

Can historical search and engagement metrics be used to identify content pages that are likely to decline in search performance?

## Unit of Analysis

One content page.

## Expected Output

A binary prediction:

- Declining
- Not Declining

## Decision

Which pages should an SEO team review first?

## Action

Pages predicted to decline can be prioritized for content updates, keyword optimization, or technical improvements.

## Cost of an Incorrect Recommendation

### False Positive

A page predicted to decline may actually remain stable, causing unnecessary effort.

### False Negative

A page that is actually declining may not be reviewed in time, potentially leading to reduced traffic and missed opportunities.

Machine learning can assist in prioritizing work but should be used only as decision support rather than making automatic decisions.

```python
import pandas as pd

# Load dataset
df = pd.read_csv("../../data/raw/content_refresh_anonymized.csv")

print("Dataset Shape:", df.shape)

print("\nFirst Five Rows")
display(df.head())

print("\nDeclining Label Distribution")
print(df["is_declining_label"].value_counts())

print("\nAverage CTR (%)")
print(round(df["ctr"].mean(),2))

print("\nAverage Search Position")
print(round(df["avg_position"].replace(0,pd.NA).mean(),2))

print("\nAverage Engagement Rate (%)")
print(round(df["engagement_rate"].mean(),2))
```
